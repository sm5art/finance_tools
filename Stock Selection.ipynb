```python
import pandas as pd
```

```python
def nasdaq_list():
    df = pd.read_csv('companylist.csv')
    return df
```

```python
company_list = nasdaq_list()
```

```python
prices_under_5 = company_list[company_list['LastSale'] < 5]
```

```python
market_caps = prices_under_5['MarketCap'].fillna("$0.0M").apply(lambda x: x[1:-1]).astype(float)
```

```python
prices_under_5['MarketCap'] = market_caps
```

```python
sorted_by_market_cap = prices_under_5.sort_values('MarketCap', ascending=False)
```
