```python
import pandas as pd
```

```python
year = '2019'
month = 'Jan'
day = '22'
```

```python
def get_df(month, year, day):
    earnings = pd.read_html('https://www.nasdaq.com/earnings/earnings-calendar.aspx?date=%s-%s-%s'%(year, month, day))
    return earnings[0]

def date_range(month, year, start, delta):
    earnings_df = get_df(month, year, start)
    for i in range(delta-1):
        start = str(int(start) + 1)
        earnings_df = earnings_df.append(get_df(month, year, start))
    return earnings_df
```

```python
earnings = date_range(month, year, day, 3)
```
