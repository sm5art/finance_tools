```python
import pandas as pd
```

```python
year = '2019'
month = 'March'
day = '1'
```

```python
def get_df(month, year, day):
    earnings = pd.read_html('https://www.nasdaq.com/earnings/earnings-calendar.aspx?date=%s-%s-%s'%(year, month, day))
    return earnings[0]

def date_range(month, year, start, delta):
    earnings_df = get_df(month, year, start)
    for i in range(delta-1):
        start = str(int(start) + 1)
        earnings_df = earnings_df.append(get_df(month, year, start))
    return earnings_df
```

```python
earnings = date_range(month, year, day, 5)
```

```python
earnings
```

```python
import re
pattern = re.compile(r"\(([A-Z]+)\)")
```

```python
earnings.columns[4]
```

```python
def filter(wow):
    results = []
    if type(wow) is str:
        results = pattern.findall(wow)
    if len(results) > 0:
        return results[0]
    else:
        return None
```

```python
tickers = earnings[earnings.columns[4]].apply(filter)

earnings['ticker'] = tickers
```

```python
tickers
```

```python
def nasdaq_list():
    df = pd.read_csv('companylist.csv')
    return df
```

```python
company_list = nasdaq_list()
```

```python
prices_under_5 = company_list[company_list['LastSale'] < 5]
```
