```python
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
import matplotlib.pyplot as plt
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
def get_data_for_symbol(symbol):
    return pd.read_csv("http://www.stockpup.com/data/%s_quarterly_financial_data.csv" % symbol)
```

```python
data = get_data_for_symbol("AAPL")

data = data[::-1]
```

```python
date_index = data['Quarter end']
```

```python
plot_types = {"line": plt.plot, "scatter": plt.scatter}

def plot(column, typ):
    plot_types[typ](date_index, data[column])
    plt.show()
```

```python
plot('Price', 'line')
```

```python
plot('Asset turnover', 'line')
```

```python

def correlations(col_start, col_end):
    cols = list(data.columns[col_start: col_end].drop(['Quarter end', 'Price'], errors="ignore"))

    float_cols = data[cols].replace('None', np.nan).astype(float, errors='ignore').iloc[:-1].reset_index(drop=True)
    float_cols = float_cols.fillna(0)

    returns = data['Price'].replace('None', np.nan).astype(float, errors='ignore').iloc[1:].reset_index(drop=True)
    float_cols['Returns'] = returns
    cols += ['Returns']
    float_cols

    correlations = float_cols.pct_change().corr()
    # plot correlation matrix
    fig = plt.figure()
    names = []
    ax = fig.add_subplot(111)
    cax = ax.matshow(correlations, vmin=-1, vmax=1)
    fig.colorbar(cax)
    ticks = np.arange(0,len(cols),1)
    ax.set_xticks(ticks)
    ax.set_yticks(ticks)
    ax.set_xticklabels(cols)
    ax.set_yticklabels(cols)
    plt.show()
```

```python
correlations(0, 10)
```
